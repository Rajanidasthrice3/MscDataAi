```python
# Combining different datasets is a common task in data analysis. Pandas
# provides several ways to combine DataFrames, with merging being one of the
# most powerful and flexible. Merging allows you to combine DataFrames based on
# common columns or indices, similar to how you would join tables in a
# relational database.
#
# The primary function for merging DataFrames is pd.merge(). This function
# offers various types of merges, controlled by the how parameter:
#
# inner (default): Returns only the rows where the merge key(s) are present in
# both DataFrames. This is the intersection of the two DataFrames' keys.
#
# left: Returns all rows from the left DataFrame and the matched rows from the
# right DataFrame. Missing values (NaN) will appear in the columns from the
# right DataFrame for rows that don't have a match.
#
# right: Returns all rows from the right DataFrame and the matched rows from
# the left DataFrame. Missing values (NaN) will appear in the columns from the
# left DataFrame for rows that don't have a match.
#
# outer: Returns all rows when there is a match in either the left or the right
# DataFrame. This is the union of the two DataFrames' keys. Missing values (NaN)
# will appear where there is no match.
#
# Let's create two sample DataFrames and demonstrate these different merge types

# 1. Create two sample DataFrames
data1 = {'ID': [1, 2, 3, 4],
         'Name': ['Alice', 'Bob', 'Charlie', 'David'],
         'Age': [25, 30, 35, 40]}
df1 = pd.DataFrame(data1)

data2 = {'ID': [1, 3, 4, 5],
         'City': ['New York', 'London', 'Paris', 'Tokyo'],
         'Salary': [50000, 60000, 70000, 80000]}
df2 = pd.DataFrame(data2)

print("DataFrame 1:")
display(df1)

print("\nDataFrame 2:")
display(df2)

```

```python
# 2. Perform an inner merge
print("\nInner Merge:")
inner_merge_df = pd.merge(df1, df2, on='ID', how='inner')
display(inner_merge_df)
```

```python

# 3. Perform a left merge
print("\nLeft Merge:")
left_merge_df = pd.merge(df1, df2, on='ID', how='left')
display(left_merge_df)

```

```python
# 4. Perform a right merge
print("\nRight Merge:")
right_merge_df = pd.merge(df1, df2, on='ID', how='right')
display(right_merge_df)

```

```python
# 5. Perform an outer merge
print("\nOuter Merge:")
outer_merge_df = pd.merge(df1, df2, on='ID', how='outer')
display(outer_merge_df)
```
