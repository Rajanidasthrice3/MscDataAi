```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,8*np.pi, 1000)
y=np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y,color='black',linestyle='--',linewidth=1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A Sine Wave')
plt.show()
```

```python
x=np.linspace(0,10, 10)
y=np.cos(x)
fig, ax = plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A Sine Wave')
ax.plot(x,y,color='black',marker='o')
plt.show()
```

```python
x=np.random.uniform(0,1,100)
y=np.random.uniform(0,1,100)
fig, ax = plt.subplots()
ax.scatter(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A Scatterplot')
plt.show()
fig.savefig('scatter.png', dpi=300)
```

```python
x=np.arange(12)
y=np.random.uniform(0,12,12)
fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_xlabel('Number of Men')
ax.set_ylabel('Salary (in lakhs)')
ax.set_title('A Bar graph')
plt.show()
```

```python
z=np.random.uniform(0,1,(8,8))
fig, ax = plt.subplots()
ax.set_xlabel('Number of Men')
ax.set_ylabel('Salary (in lakhs)')
ax.set_title('A Heat graph')
ax.imshow(z)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
z=np.random.uniform(1,2,4)
fig, ax = plt.subplots()
ax.set_title('A Pie graph')
ax.pie(z, labels=['A','B','C','D'], autopct='%1.1f%%')
ax.legend(loc='upper right')
plt.show()
```

```python
z=np.random.normal(0,1,50)
fig, ax = plt.subplots()
ax.hist(z, 15)
plt.show()
```
